```python
df = pd.read_csv('weather_data.csv')
df
```

```python
#replacing -99999 with NA

new_df = df.replace(-99999,np.NaN)
new_df
```

```python
#say we have -88888 value also

df = pd.read_csv('weather_data.csv')
df
```

```python
new_df = df.replace([-99999,-88888],np.NaN)#supplying a list
new_df
```

```python
df = pd.read_csv('weather_data.csv')
df
```

```python
'''
Here inside event we would like to replace 0 as well but doing 
  
  new_df = df.replace([-99999,-88888,0],np.NaN)

will replace 0 in temperature & windspeed and 0 is a valid value in this column. So, passing a dictionary 
'''

new_df = df.replace({
    'temperature':-99999,
    'windspeed':[-99999,-88888],
    'event':'0'
},np.NaN)#supplying a dictionary
new_df
```

```python
df = pd.read_csv('weather_data.csv')
df
```

```python
#now we would like to replace -99999 with NaN and 'No Event' with 'Sunny'

new_df = df.replace({
    -99999:np.NaN,
    -88888:np.NaN,
    'No Event':'Sunny'
})
new_df
```

```python
df = pd.read_csv('weather_data.csv')
df
```

```python
'''
Here we should remove mph and C, F as their unit. The bet way to remove such values is to use regular expression to 
detect pattern.
'''

new_df = df.replace('[A-Za-z]','',regex=True)
new_df
```

```python
#The problem is it removed event as well so using dictionary

new_df = df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]',
    },'',regex=True)

new_df
```
