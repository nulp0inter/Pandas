```python
#https://www.youtube.com/watch?v=WGOEFok1szA&list=PLeo1K3hjS3uuASpe-1LjfG5f14Bnozjwy&index=8
#https://pandas.pydata.org/docs/reference/api/pandas.concat.html

import pandas as pd
```

```python
india_weather = pd.DataFrame({
    'city':['mumbai','delhi','banglore'],
    'temperature':[32,45,30],
    'humidity':[80,60,78]
})

india_weather
```

```python
us_weather = pd.DataFrame({
    'city':['new york','chicago','orlando'],
    'temperature':[21,14,35],
    'humidity':[68,65,75]
})

us_weather
```

```python
#concating the dataframe

df = pd.concat([india_weather,us_weather])
df
```

```python
#notice that the index is not continuous so for that

df = pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
#setting index in database

df = pd.concat([india_weather,us_weather],keys=['india','usa'])
df
```

```python
df.loc['usa']#accessing by index
```

```python
#concating columnwise data

temperature_df = pd.DataFrame({
    'city':['mumbai','delhi','banglore'],
    'temperature':[32,45,30]
})

temperature_df
```

```python
windspeed_df = pd.DataFrame({
    'city':['mumbai','delhi','banglore'],
    'windspeed':[7,12,9]
})

windspeed_df
```

```python
df = pd.concat([temperature_df,windspeed_df])
df
```

```python
#However we would like to append windspeed at the end of temperature
#By default axis = 0 which will append as a row and if it is changed to 1 it appends as a column

df = pd.concat([temperature_df,windspeed_df],axis=1)
df
```

```python
#now lets change the order of city and a data is missing

windspeed_df = pd.DataFrame({
    'city':['delhi','mumbai'],
    'windspeed':[7,12]
})

windspeed_df
```

```python
df = pd.concat([temperature_df,windspeed_df],axis=1)
df
```

```python
#this doesnot look rigth so using index argument

temperature_df = pd.DataFrame({
    'city':['mumbai','delhi','banglore'],
    'temperature':[32,45,30]
},  index=[1,2,0])

temperature_df
```

```python
windspeed_df = pd.DataFrame({
    'city':['delhi','mumbai'],
    'windspeed':[7,1]
},  index = [2,1])

windspeed_df
```

```python
df = pd.concat([temperature_df,windspeed_df],axis=1)
df
```
