```python
#we will convert it to date type

df = pd.read_csv('weather_data.csv',parse_dates=['day'])
df.dtypes
```

```python
#setting date as index

df.set_index('day',inplace=True)
```

```python
df
```

```python
#filling the missing values with 0

new_df = df.fillna(0)
new_df
```

```python
#we might not always want to fill NA values with some values in all column so passing a dictionary for especific column

new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

```python
#Another way to replace missing value is to carry forward the previous vaues i.e. tem of 1/4 = tem of 1/1

new_df = df.fillna(method='ffill') #forwardfill method
new_df
```

```python
new_df = df.fillna(method='bfill') #backwardfill method
new_df
```

```python
df
```

```python
new_df = df.fillna(method='bfill',axis='columns') #backwardfill method but column wise
new_df
```

```python
new_df = df.fillna(method='ffill',limit=1) #limit = 1 sets that the previous value will be filled just once
new_df
```

```python
new_df = df.copy() #replacing NA values with mean, median and mode
new_df['temperature'].fillna(value=new_df['temperature'].mean(), inplace=True) 
new_df['windspeed'].fillna(value=new_df['windspeed'].median(), inplace=True)
new_df
```

```python
#replacing with mod
new_df['temperature'].mode()#this vaalue should be indexed with new_df['temperature'].mode()[0]
```

```python
new_df['windspeed'].mode()#here their are 4 mode values so this will caluse an error f we try to replace NA
# 
```

```python
new_df = df.copy() #replacing NA values with mean, median and mode
new_df['temperature'].fillna(value=new_df['temperature'].mode()[0], inplace=True) 
new_df['windspeed'].fillna(value=new_df['windspeed'].mode()[0], inplace=True) 
new_df
```

```python
#Now Filling NA values with a fixed value might not be the best estimation. Interpolation can be used for better estimation
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.interpolate.html

new_df=df.interpolate() #this is a linear interpolation
new_df     

'''
different methods for interpolation are 

    ‘linear’     ‘time’     ‘index’    ‘values’     ‘pad’     ‘nearest’     ‘zero’     ‘slinear’     ‘quadratic’ 
    ‘cubic’     ‘barycentric’      ‘polynomial’    ‘krogh’    ‘piecewise_polynomial’    ‘spline’     ‘pchip’
    ‘akima’     ‘cubicspline’     ‘from_derivatives’
'''
```

```python
'''
One thing we can notice in above data is in '2017-01-04' interpolated temperature is 30 which is average of 32 & 28.
However, the date '2017-01-04' is more close to '2017-01-05' than '2017-01-01' so temperature should be near to 28 
rather than 32. To solve this we will use method='time' for interpolation
'''

new_df=df.interpolate(method='time') 
new_df 
```

```python
#Dropping all NA values

new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')#dropping NA values only if all rows are empty
new_df
```

```python
new_df = df.dropna(thresh=2)#threshold = 2, if 2 NA or more then drop column 
new_df
```

```python
#inserting missing date betwween 01-01-2017 to 01-11-2017

dt = pd.date_range('01-01-2017','01-11-2017')#creating a range
index = pd.DatetimeIndex(dt)#passing to datetime index
df = df.reindex(index)#inserting missing range
df
```
